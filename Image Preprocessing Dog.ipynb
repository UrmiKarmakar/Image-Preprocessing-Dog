```python
import matplotlib.pyplot as plt

resized_image = cv2.resize(image, (128, 128))
plt.imshow(resized_image)
plt.show()
```

```python
from imgaug import augmenters as iaa
import matplotlib.pyplot as plt
```

```python
seq=iaa.Sequential([
    iaa.Fliplr(0.5), #horizontal flip with probability 50%
    iaa.Flipud(0.5), #vertical flip with probability 50%
    iaa.Affine(rotate=(-15,15)) #random rotate
])

augmented_image=seq.augment_image(image)

plt.imshow(augmented_image)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.axis(False)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(augmented_image)
plt.axis(False)
plt.show()
```

```python
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.axis(False)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(grayscale_image, cmap='gray')
plt.axis(False)
plt.show()
```

```python
normalized_image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)
```

```python
print(image[0:5, 0:5])
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.axis(False)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(normalized_image)
plt.axis(False)
plt.show()
```

```python
rescaled_image = image / 255.0
```
